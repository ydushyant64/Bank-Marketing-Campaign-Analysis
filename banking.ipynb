```python
# Default rate by age group

default_by_age = pd.crosstab(df['age_group'], df['default'], normalize='index') * 100
print("Default Rate by Age Group (%):")
print(default_by_age.round(2))
```

```python
# Default rate by job

default_by_job = pd.crosstab(df['job'], df['default'], normalize='index') * 100
print("Default Rate by Job Type (%):")
print(default_by_job.round(2))

# print(default_by_job.columns)

# Sort by default rate
if 'yes' in default_by_job.columns:
    default_by_job_sorted = default_by_job.sort_values('yes', ascending=False)
    print("\nJobs with Highest Default Rates (%):")
    print((default_by_job_sorted['yes'].head(12).round(2)))
```

```python
# Overall default statistics

total_clients = len(df)
default_yes = len(df[df['default'] == 'yes'])
default_no = len(df[df['default'] == 'no'])
default_rate = (default_yes / total_clients) * 100

print("=== CREDIT DEFAULT SUMMARY ===\n")
print(f"Total Clients: {total_clients:,}")
print(f"Clients in Default: {default_yes:,}")
print(f"Clients NOT in Default: {default_no:,}")
print(f"Default Rate: {default_rate:.2f}%")
print(f"Non-Default Rate: {100-default_rate:.2f}%")

# Risk assessment
if default_rate < 5:
    risk_level = "Low Risk"
elif default_rate < 15:
    risk_level = "Moderate Risk"
else:
    risk_level = "High Risk"

print(f"Risk Level: {risk_level}")
```

```python
# Additional insights
print("=== KEY INSIGHTS === ")

# 1. Which job has highest default rate?
if 'yes' in default_by_job.columns:
    job_default_rates = default_by_job['yes']
    highest_default_job = job_default_rates.idxmax()
    highest_default_rate = job_default_rates.max()
    print(f"Highest default rate job: {highest_default_job} ({highest_default_rate:.2f}%)")

# 2. Age group with highest default rate
if 'yes' in default_by_age.columns:
    age_default_rates = default_by_age['yes']
    highest_default_age = age_default_rates.idxmax()
    highest_default_age_rate = age_default_rates.max()
    print(f"Highest default rate age group: {highest_default_age} ({highest_default_age_rate:.2f}%)")
```

## 6. What is the distribution of average yearly balance among the clients?

```python
df['balance'].value_counts()
```

```python
# 1. Density plot

plt.figure(figsize=(10,8))
sns.histplot(df['balance'], kde = True, bins=50, alpha=0.7)
plt.title('Balance distribution with density curve')
plt.xlabel('Balance')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['balance'])
plt.title('Balance Distribution - Box Plot')
plt.xlabel('Balance ($)')
plt.tight_layout()
plt.show()
```

```python
# Create balance categories
def categorize_balance(balance):
    if balance < 0:
        return 'Negative (Debt)'
    elif balance == 0:
        return 'Zero Balance'
    elif balance < 1000:
        return 'Low ($1-999)'
    elif balance < 5000:
        return 'Medium ($1K-5K)'
    elif balance < 20000:
        return 'High ($5K-20K)'
    else:
        return 'Very High ($20K+)'

df['balance_category'] = df['balance'].apply(categorize_balance)

# Count and percentage by balance category
balance_counts = df['balance_category'].value_counts()
balance_percentages = df['balance_category'].value_counts(normalize=True) * 100

print("==== Balance Category Distribution ====\n")
for category in balance_counts.index:
    print(f"{category}: {balance_counts[category]:,} ({balance_percentages[category]:.1f}%)")
```

```python
plt.figure(figsize=(12, 6))
balance_counts = df['balance_category'].value_counts()
# balance_counts.plot(kind='bar',color='lightblue', edgecolor='black', alpha = 0.7)
bars = plt.bar(balance_counts.index, balance_counts.values, color='lightblue', edgecolor='black')

# Add percentage labels on bars
total = len(df) 
for bar, count in zip(bars, balance_counts.values):
    percentage = (count/total) * 100
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 50,f'{percentage:.1f}%', ha='center', va='bottom', fontweight='bold')

plt.title('Distribution of Clients by Balance Category')
plt.xlabel('Balance Category')
plt.ylabel('Number of Clients')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Identify outliers using IQR method
Q1 = df['balance'].quantile(0.25)
Q3 = df['balance'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['balance'] < lower_bound) | (df['balance'] > upper_bound)]
print(f"====== Outlier Analysis ====== \n")
print(f"Lower bound: ${lower_bound:,.2f}")
print(f"Upper bound: ${upper_bound:,.2f}")
print(f"Number of outliers: {len(outliers):,} ({len(outliers)/len(df)*100:.1f}%)")
print(f"Highest balance: ${df['balance'].max():,.2f}")
print(f"Lowest balance: ${df['balance'].min():,.2f}")
```

```python
# Average balance by job
balance_by_job = df.groupby('job')['balance'].agg(['mean', 'median', 'std']).round(2)
balance_by_job = balance_by_job.sort_values('mean', ascending=False)

print("Balance Statistics by Job Type:\n")
print(balance_by_job)

# Visualization
plt.figure(figsize=(12, 8))
balance_by_job['mean'].plot(kind='bar', color='lightblue', edgecolor='black')
plt.title('Average Balance by Job Type')
plt.xlabel('Job Type')
plt.ylabel('Average Balance ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Balance by age group
balance_by_age = df.groupby('age_group')['balance'].agg(['mean', 'median', 'count']).round(2)
print("Balance by Age Group:")
print(balance_by_age)

# Box plot by age group
plt.figure(figsize=(12, 6))
sns.boxplot(data=df, x='age_group', y='balance')
plt.title('Balance Distribution by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Balance ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Remove extreme outliers for better visualization
balance_filtered = df[(df['balance'] >= df['balance'].quantile(0.01)) & (df['balance'] <= df['balance'].quantile(0.99))]

plt.figure(figsize=(12, 6))
plt.hist(balance_filtered['balance'], bins=50, edgecolor='black', alpha=0.7, color='lightcoral')
plt.title('Balance Distribution (1st-99th Percentile)')
plt.xlabel('Balance ($)')
plt.ylabel('Frequency')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Percentile breakdown
percentiles = [10, 25, 50, 75, 90, 95, 99]
print("Balance Percentiles:")
for p in percentiles:
    value = df['balance'].quantile(p/100)
    print(f"{p}th percentile: ${value:,.2f}")

# What percentage have negative balance?
negative_balance = (df['balance'] < 0).sum()
negative_percentage = (negative_balance / len(df)) * 100
print(f"\nClients with negative balance: {negative_balance:,} ({negative_percentage:.1f}%)")

# What percentage have zero balance?
zero_balance = (df['balance'] == 0).sum()
zero_percentage = (zero_balance / len(df)) * 100
print(f"Clients with zero balance: {zero_balance:,} ({zero_percentage:.1f}%)")
```

```python
# Create duration categories
def categorize_duration(duration):
    if duration == 0:
        return 'No Contact (0 sec)'
    elif duration <= 60:
        return 'Very Short (1-60 sec)'
    elif duration <= 180:
        return 'Short (61-180 sec)'
    elif duration <= 300:
        return 'Medium (181-300 sec)'
    elif duration <= 60:
        return 'Long (301-600 sec)'
    else:
        return 'Very Long (600+ sec)'

df['duration_category'] = df['duration'].apply(categorize_duration)

# Count and percentage analysis
duration_counts = df['duration_category'].value_counts()
duration_percentages = df['duration_category'].value_counts(normalize=True) * 100

print("Duration Categories Distribution:")
for category in duration_counts.index:
    count = duration_counts[category]
    percentage = duration_percentages[category]
    print(f"{category}: {count} ({percentage :0.3}%)")
```

```python
# bar plot for duration categories

plt.figure(figsize=(12, 6))
bars = plt.bar(duration_counts.index, duration_counts.values, color='steelblue', alpha=0.8)
plt.title('Distribution of Contact Duration Categories', fontsize=16, fontweight='bold')
plt.xlabel('Duration Categories', fontsize=12)
plt.ylabel('Number of Contacts', fontsize=12)
plt.xticks(rotation=45, ha='right')

# Adding percentage labels on bars
total = len(df)
for bar, count in zip(bars, duration_counts.values):
    percentage = (count/total) * 100
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 50,
             f'{percentage:.3f}%', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()
```
